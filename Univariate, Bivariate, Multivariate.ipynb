```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
# The data itself (the features) can be accessed using iris.data. This is a NumPy array where each row 
# corresponds to an iris flower and each column 
# corresponds to a feature (sepal length, sepal width, petal length, petal width).
iris
```

```python
iris_data=iris.data
```

```python
iris_data
```

```python
# The names of the features can be accessed using iris.feature_names. 
# This is a list containing the names of the features in the same order as the columns in iris.data.
iris_feature_names=iris.feature_names
```

```python
iris_feature_names
```

```python
df=pd.DataFrame(iris_data, columns=iris_feature_names)
```

```python
df
```

```python
df["species"]=iris.target
```

```python
# iris.target_names is an array of species names: ['setosa', 'versicolor', 'virginica'].
# enumerate(iris.target_names) generates pairs of index and species names: [(0, 'setosa'), (1, 'versicolor'), (2, 'virginica')].
# {i: name for i, name in enumerate(iris.target_names)} creates a dictionary mapping numerical labels to species names: {0: 'setosa', 1: 'versicolor', 2: 'virginica'}.
# iris_df['species'].map({...}) maps each numerical label in the species column to its corresponding species name using the dictionary created.
# iris_df['species_name'] creates a new column in the DataFrame and assigns the mapped species names to it.

df["species_names"]=df["species"].map({i:name for i, name in enumerate(iris.target_names)})
```
