```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
df=sns.load_dataset('iris')
```

```python
df.head(5)
```

```python
# independent and independent features
X=df.iloc[:,:4]
```

```python
X
```

```python
y=iris.target
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# postpruning
treemodel=DecisionTreeClassifier()
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
# postpruning(above tree is the complete tree we got but we need only 2 levels, so mention max depth level=2)
# if parameter not mentioned then whole tree is displayed
treemodel=DecisionTreeClassifier(max_depth=2)
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
y_pred=treemodel.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
classification_report=classification_report(y_pred,y_test)
```

```python
print("score:",score,"\n" "classification_report:",classification_report)
```

```python
# prepruning-pruning is done while building model
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
treemodel=DecisionTreeClassifier()
cv=GridSearchCV(treemodel,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred=cv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
score=accuracy_score(y_pred,y_test)
classification_report=classification_report(y_pred,y_test)
```
