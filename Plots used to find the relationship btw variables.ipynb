```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# 1.scatter plot- relationship btw 2 continuous vars
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 2.lineplot-connects datapoints using lines and used in time series data or showing trends over interval

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 3.pair plot- shows realtion btw every feature, includes histogram and density plots in diagonal
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [2, 3, 5, 7, 11],
    'C': [1, 4, 9, 16, 25]
}
df = pd.DataFrame(data)

# Pair plot
sns.pairplot(df)
plt.show()
```

```python
# 4.heatmap- uses color to represent the values of the variable, effective to show correlation 

data = np.random.rand(10, 10)
df = pd.DataFrame(data)

# Heatmap
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlations')
plt.show()
```

```python
# 5.joint plot- Combines scatter plots, histograms or density plots to show the 
# relationship between two variables and their distributions.
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Joint plot
sns.jointplot(x=x, y=y, kind='scatter')
plt.show()
```

```python
# 6.regression line-plots a line to show linear relationship btw variables

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Regression plot
sns.regplot(x=x, y=y, scatter_kws={'color': 'blue'}, line_kws={'color': 'green'})
plt.title('Regression Plot')
plt.show()
```
