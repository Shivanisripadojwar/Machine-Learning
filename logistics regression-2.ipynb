```python
# Outlier Identification
outliers_column=['cigsPerDay', 'BPMeds','prevalentStroke', 'diabetes', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose', 'TenYearCHD']
```

```python
# calculate number of outliers based on IQR
def calculate_outliers(df,column):
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    outliers=df[(df[column]<lower_bound)|(df[column]>upper_bound)]
    return outliers
outliers_dict={}
outliers_count={}
for column in outliers_column:
    outliers=calculate_outliers(df,column)
    outliers_dict[column]=outliers
    outliers_count[column]=outliers.shape
```

```python
for column,outliers in outliers_dict.items():
    print(f"total number of outliers in '{column}': {outliers_count[column]}")
    print(f"outliers in '{column}':")
    print(outliers)
    print() 
```

```python
# model training
X=df.drop('TenYearCHD',axis=1)
Y=df['TenYearCHD']
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
logistic_model=LogisticRegression()
logistic_model.fit(x_train,y_train)
```

```python
# model evaluation
y_test
```

```python
y_pred=logistic_model.predict(x_test)
```

```python
y_pred
```

```python
plt.figure(figsize=(5,5))
sns.histplot(y_pred,kde=True,color='blue',label='pred values')
sns.histplot(y_test,kde=True,color='red',label='actual values')
plt.xlabel('values')
plt.title("actual vs predicted values")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
print("confusion_matrix",confusion_matrix(y_test,y_pred))
print("accuracy_score",accuracy_score(y_test,y_pred))
print("classification_report",classification_report(y_test,y_pred))
```

```python
# plot confusion matrix
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,cmap='Blues',fmt='d')
plt.title('confusion matrix')
plt.ylabel('actual')
plt.xlabel('predicted')
plt.show()
```
